```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import ipywidgets as widgets
from IPython.display import display
```

```python
data='https://docs.google.com/spreadsheets/d/e/2PACX-1vRVA-3I_H9KCh-HH3ec7dkfrWYN0xO0Onu1-1OtxuajBv1qT4GDOrt8YwVDtur5OFSM1iolZ9UV8JBB/pub?gid=0&single=true&output=csv'
```

```python
df=pd.read_csv(data)
```

```python
df
```

```python
df=df.drop('Id',axis=1)
```

```python
np.array(df)
```

```python
y=df.drop(['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'],axis=1)
```

```python
y
```

```python
X=df.drop('Species',axis=1)
```

```python
X
```

```python
X=np.array(X)
```

```python
X
```

```python
y=np.array(y)
y
```

```python

label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
y
```

```python
model=Sequential([Dense(units=15,activation='relu'),
                  Dense(units=10,activation='relu'),
                  Dense(units=3,activation='linear')])
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.4,random_state=42)
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=100)
```

```python
loss,accu=model.evaluate(x_test,y_test)
```

```python
loss_train,accuracy_train=model.evaluate(x_train,y_train)
```
